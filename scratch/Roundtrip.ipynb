```python
import librosa
%matplotlib inline
import IPython.display as ipd
import librosa.display
```

```python
input_path = '../samples/steve-morrell/giveit-07.wav'
output_path = 'test-out.wav'
```

```python
ipd.Audio(input_path)
```

```python
x, sr = librosa.load(input_path)
librosa.display.waveplot(x, sr=sr)
```

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(X)
```

```python
X
```

```python
X2 = librosa.db_to_amplitude(Xdb)
```

```python
x2 = librosa.core.istft(X)
```

```python
librosa.output.write_wav(output_path, x2, sr)
librosa.display.waveplot(x2, sr=sr)
```

```python
ipd.Audio(output_path)
```
